```python
import numpy as np
import seaborn as sns; sns.set(color_codes=True)
from keras.models import Model, Sequential
from keras.layers import Input, Dense
from matplotlib import pyplot as plt
```

## Build Model

```python
model0 = Sequential()
model0.add(Dense(5, activation="relu", input_dim=1, name="h1"))
model0.add(Dense(10, activation="relu", name="h2"))
model0.add(Dense(10, activation="relu", name="h3"))
model0.add(Dense(10, activation="relu", name="h4"))
model0.add(Dense(10, activation="relu", name="h5"))
model0.add(Dense(10, activation="relu", name="h6"))
model0.add(Dense(5, activation="relu", name="h7"))
model0.add(Dense(1, name="h8"))

model0.summary()
```

```python
model1 = Sequential()
model1.add(Dense(10, activation="relu", input_dim=1, name="h1"))
model1.add(Dense(18, activation="relu", name="h2"))
model1.add(Dense(15, activation="relu", name="h3"))
model1.add(Dense(4, activation="relu", name="h4"))
model1.add(Dense(1, name="h5"))

model1.summary()
```

```python
model2 = Sequential()
model2.add(Dense(190, activation="relu", input_dim=1, name="h1"))
model2.add(Dense(1, name="h2"))

model2.summary()
```

## Generate Data 1

```python
train_x = np.random.uniform(low=1e-9, high=1, size=(10000,))
test_x = np.linspace(0, 1, num=100)
```

```python
# Add a small number to prevent dividing zero
test_x[0] += 1e-9
```

```python
train_y = np.sin(5*np.pi*train_x) / (5*np.pi*train_x)
test_y = np.sin(5*np.pi*test_x) / (5*np.pi*test_x)
```

```python
idx = np.argsort(train_x)
```

```python
# Plot
plt.title("Training Data")
plt.plot(train_x[idx], train_y[idx])
plt.show()
plt.title("Testing Data")
plt.plot(test_x, test_y)
plt.show()
```

## Training

```python
model0.compile(optimizer="adam", loss="mse")
history0 = model0.fit(train_x, train_y, batch_size=256, epochs=1000, validation_data=(test_x, test_y))
```

```python
model1.compile(optimizer="adam", loss="mse")
history1 = model1.fit(train_x, train_y, batch_size=256, epochs=1000, validation_data=(test_x, test_y))
```

```python
model2.compile(optimizer="adam", loss="mse")
history2 = model2.fit(train_x, train_y, batch_size=256, epochs=1000, validation_data=(test_x, test_y))
```

## Plot

```python
# Plot Loss
plt.title("Learning Curve of model0")
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.plot(history0.history["loss"], label="Training")
plt.plot(history0.history["val_loss"], label="Testing")
plt.legend()
plt.show()
plt.title("Learning Curve of model1")
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.plot(history1.history["loss"], label="Training")
plt.plot(history1.history["val_loss"], label="Testing")
plt.legend()
plt.show()
plt.title("Learning Curve of model2")
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.plot(history2.history["loss"], label="Training")
plt.plot(history2.history["val_loss"], label="Testing")
plt.legend()
plt.show()
```

```python
preds0 = model0.predict(test_x, batch_size=256)
preds1 = model1.predict(test_x, batch_size=256)
preds2 = model2.predict(test_x, batch_size=256)
```

```python
# Plot Predictions
plt.plot(test_x, test_y, label="Ground Truth")
plt.plot(test_x, preds0, label="model0")
plt.plot(test_x, preds1, label="model1")
plt.plot(test_x, preds2, label="model2")
plt.legend()
plt.show()
```
